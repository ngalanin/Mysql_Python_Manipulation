```python
import sqlite3
import pandas as pd
import numpy as np

```

```python
conn = sqlite3.connect('Immobilien.db3')

print ("Opened database successfully")
```

```python
cur = conn.cursor()
```

```python
df = pd.read_sql_query("select * from Mietern;", conn,coerce_float=True)
df
```

```python
df.dtypes
```

```python
date2=pd.to_datetime(df.Today)
date1=pd.to_datetime(df.Startdate)

df['Monaten'] = ((date2 - date1)/np.timedelta64(1, 'M'))


df['Monaten'] = df['Monaten'].astype(int)
```

```python
df['Monaten']
```

```python
df
```

```python
df["Kaltmiete"] = df["Kaltmiete"].astype(str).astype(float)
df['Saldo']=df.Monaten*(df.Kaltmiete+df.Nebenkosten)
```

```python
df
```

```python
bezahlt=mayer['Betrag'].sum()

soll=df.loc[0].at['Saldo']
df.at[0, 'Saldo'] = bezahlt-soll
df
```
